```python
# Check whether there are colums with mixed data types:

def cols_with_mixed_data_types(df):
# function for determining if there are multiple data types in a column:
    for column in df.columns:
        if len(df[column].map(type).unique())>1:
            print(f"{column}: {df[column].map(type).unique()}")

# call the function:
cols_with_mixed_data_types(df)
```
